```python
#ajuste das datas
acoes_2021['data'] = pd.to_datetime(acoes_2021['data'])
```

```python
#Achando as empresas com maiores preço de abertura médio 
media_precos_2021_empresa = acoes_2021.groupby('empresa')['preco_abertura'].mean().round(4).sort_values(ascending=False)
media_precos_2021_empresa.head(20)
```

```python
#escolha de carteira com 7 ativos
carteira_empresas = [ 'ABEV3', 'ARZZ3', 'BBSE3', 'BBDC4', 'BBAS3', 'BRKM5', 'CIEL3']
```

```python
#análise de cada empresa na carteira
for empresa in carteira_empresas:
     df_empresa = acoes_2021[(acoes_2021.cod == empresa)].sort_values(by = 'data')
     print(df_empresa)
     plt.plot(df_empresa['data'], df_empresa['preco_abertura'], label= empresa)
     plt.title('Variação do Preço ao Longo das Datas')
     plt.xlabel('Data')
     plt.ylabel('Preço')
     plt.legend()
     plt.show()
```

```python
#definindo um data frame para resumir as informações da carteira para este ano
df_2 = pd.DataFrame(carteira_empresas, columns=['cod'])
df_2
```

```python
df_2 = pd.merge(df_2, acoes_2021[acoes_2021['data'] == '2021-01-04'][['cod', 'preco_abertura']], how='left', left_on='cod', right_on='cod', suffixes=('', '_inicial'))
df_2 = pd.merge(df_2, acoes_2021[acoes_2021['data'] == '2021-12-30'][['cod', 'preco_abertura']], how='left', left_on='cod', right_on='cod', suffixes=('', '_inicial'))
df_2.columns = ['cod', 'preco_inicial', 'preco_final']
df_2
```

```python
df_2['retorno'] = (df_2['preco_final'] - df_2['preco_inicial'])/ df_2['preco_inicial']
df_2
```

```python
media_por_cod = acoes_2021.groupby('cod')['preco_abertura'].mean().reset_index()
df_2 = pd.merge(df_2, media_por_cod, how='left', left_on='cod', right_on='cod', suffixes=('', '_media'))
df_2 = df_2.rename(columns={'preco_abertura': 'preco_medio'})
df_2
```

```python
variancia_por_cod = acoes_2021.groupby('cod')['preco_abertura'].var().reset_index()
df_2 = pd.merge(df_2, variancia_por_cod, how='left', on='cod', suffixes=('', '_variancia'))
df_2 = df_2.rename(columns={'preco_abertura': 'variancia'})
df_2
```

```python
df_2.to_csv('df2.csv', columns= ['cod', 'retorno', 'variancia'], index= False)
```
