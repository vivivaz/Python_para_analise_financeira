```python
tamanhos =[2,8,2,12,3,12,10,3,4,11,11,11,11,11,11,11,5,18,16,11,1,8,7,7,12,3]
acoes_2022 = pd.read_fwf(arquivo, widths=tamanhos, header =0)
```

```python
#retirando a última linha (tralher)
linha = len(acoes_2022)
acoes_2022.drop(linha -1)
```

```python
#ajuste das datas
acoes_2022['data'] = pd.to_datetime(acoes_2022['data'])
```

```python
acoes_2022.isna().sum()
```

```python
#Achando as empresas com maiores preço de abertura médio 
media_precos_2022_empresa = acoes_2022.groupby('empresa')['preco_abertura'].mean().round(4).sort_values(ascending=False)
media_precos_2022_empresa.head(20)
```

```python
#escolha de carteira com 7 ativos
carteira_empresas = [ 'ABEV3', 'ARZZ3', 'BBSE3', 'BBDC4', 'BBAS3', 'BRKM5', 'CIEL3']
```

```python
#análise de cada empresa na carteira
for empresa in carteira_empresas:
     df_empresa = acoes_2022[(acoes_2022.cod == empresa)].sort_values(by = 'data')
     print(df_empresa)
     plt.plot(df_empresa['data'], df_empresa['preco_abertura'], label= empresa)
     plt.title('Variação do Preço ao Longo das Datas')
     plt.xlabel('Data')
     plt.ylabel('Preço')
     plt.legend()
     plt.show()
            
```

```python
#definindo um data frame para resumir as informações da carteira para este ano
df_1 = pd.DataFrame(carteira_empresas, columns=['cod'])
df_1
```

```python
df_1 = pd.merge(df_1, acoes_2022[acoes_2022['data'] == '2022-01-03'][['cod', 'preco_abertura']], how='left', left_on='cod', right_on='cod', suffixes=('', '_inicial'))
df_1 = pd.merge(df_1, acoes_2022[acoes_2022['data'] == '2022-12-29'][['cod', 'preco_abertura']], how='left', left_on='cod', right_on='cod', suffixes=('', '_inicial'))
df_1.columns = ['cod', 'preco_inicial', 'preco_final']
df_1
```

```python
df_1['retorno'] = (df_1['preco_final'] - df_1['preco_inicial'])/ df_1['preco_inicial']
df_1
```

```python
media_por_cod = acoes_2022.groupby('cod')['preco_abertura'].mean().reset_index()
df_1 = pd.merge(df_1, media_por_cod, how='left', left_on='cod', right_on='cod', suffixes=('', '_media'))
df_1 = df_1.rename(columns={'preco_abertura': 'preco_medio'})
df_1
```

```python
variancia_por_cod = acoes_2022.groupby('cod')['preco_abertura'].var().reset_index()
df_1 = pd.merge(df_1, variancia_por_cod, how='left', on='cod', suffixes=('', '_variancia'))
df_1 = df_1.rename(columns={'preco_abertura': 'variancia'})
df_1
```

```python
df_1.to_csv('df1.csv', columns= ['cod', 'retorno', 'variancia'], index= False)
```
